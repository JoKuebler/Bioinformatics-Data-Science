#### Subsubtask 1.3.1: Implement a ratio intensity plot (using matplotlib)

```python
# Intensity Ratio plot via Scatterplot
def ratio_plot(col_one, col_two):
    # Axes
    x = (gene_df[col_one] + gene_df[col_two])/2
    y = gene_df[col_one] - gene_df[col_two]
    # Plot 
    plt.figure(figsize=(20,15))
    plt.scatter(x, y)
    # 
    low = lowess(y[::20], x[::20], frac= 1./3, it=0)
    plt.plot([min(x), max(x)], [0,0], color='yellow')
    plt.plot(low[:,0], low[:,1], color='red')
```
