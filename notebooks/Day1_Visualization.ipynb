### A Simple Sine/Cosine plot

- Create a numpy array X with 256 values ranging from -Pi to +Pi 

```python
# Define pi and create array wit 256 values
pi = math.pi
pi_minus = -math.pi
X = np.linspace(pi_minus, pi, num=256)
```

- Create two arrays C,S with values for the sine/cosine of X

```python
# Sine and cosine
C = np.sin(X)
S = np.cos(X)
```

- Plot the curves using defaults

```python
# Plot Sine and Cosine
plt.plot(X,C)
plt.plot(X,S)
```

- Change colors and line widths

```python
# Change colors and line width
plt.plot(X,C, linewidth=7.0, c="g")
plt.plot(X,S, linewidth=0.5, c="b")
```

## Exercise: Plot the Normal Distribution

#### Task 1: Reproduce the following image as closely as you can:

<img src='https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/Normal_Distribution_PDF.svg/720px-Normal_Distribution_PDF.svg.png'>

```python
# Parameters for normal distributions
params = [(0,0.2,'blue'), (0, 1.0, 'red'), (0, 5.0, 'green'), (-2,0.5,'yellow')]

for mean, variance, color in params:
    x = np.linspace(mean - 3*math.sqrt(variance), mean + 3*math.sqrt(variance), 100)
    label = '$\mu = %.1f, \ \sigma=%.1f$' %(mean, variance)
    plt.plot(x,mlab.normpdf(x, mean, math.sqrt(variance)), label=label)

plt.xlim(-5,5)
plt.ylim(0,1)
plt.xlabel('X')
plt.ylabel('φ μ,σ(X)')
plt.legend()
plt.show()

```

#### Task 2: Now plot the above distributions as Boxplots

```python
coll_1 = np.random.normal(0, np.sqrt(0.2), 1000)
coll_2 = np.random.normal(0, np.sqrt(1.0), 1000)
coll_3 = np.random.normal(0, np.sqrt(5.0), 1000)
coll_4 = np.random.normal(-2, np.sqrt(0.5), 1000)

data = [coll_1, coll_2, coll_3, coll_4]

# Create a figure instance
fig = plt.figure(1, figsize=(9, 6))

# Create an axes instance
ax = fig.add_subplot(111)

# Create the boxplot
bp = ax.boxplot(data)
plt.show()
```
